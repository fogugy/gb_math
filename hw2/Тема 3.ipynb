#### 1. Задание (в программе)
Нарисуйте график функции: y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
def f(x, k=1, a=0, b=0):
    return k * np.cos(x-a) + b
```

#### 3.1 Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_cart(r, angle):
    # angle in radians
    x = r * np.cos(angle)
    y = r * np.sin(angle)
    return(x, y)

polar_to_cart(1, np.pi)
```

#### 3.2 Задание (в программе)
Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def cart_to_polar(x, y):
    r = np.sqrt(x**2 + y**2)
    a = np.arctan2(y, x)
    return(a, r)
```

```python
def draw_circle_in_polar(r=1):
    angle = np.linspace(0, 2*np.pi, 360) 
    plt.polar(angle, r+np.zeros(np.size(angle)))
    
draw_circle_in_polar(1.4)
```

#### 4.1 Задание (в программе)
Решите систему уравнений:
y = x**2 – 1; 
exp(x) + x∙(1 – y) = 1

```python
point_count = 200
x = np.append(np.linspace(-2, 0.00001, point_count//2), np.linspace(0.00001, 4.5, point_count//2))
y = np.linspace(-2, 20, point_count)

x0 = np.array([0]*point_count)
y0 = np.array([0]*point_count)

y1 = x**2 - 1
y2 = np.exp(x)/x + 1 -1/x

plt.plot(x,y1, color="green")
plt.plot(x,y2, color="blue")

plt.plot(x,y0, color="red", linestyle='--', linewidth=1)
plt.plot(x0,y, color="red",linestyle='--', linewidth=1)

plt.grid(True)
```

```python
def equations(p):
    x, y = p
    return (x**2-1-y, exp(x)+x*(1-y)-1)

chunks = [(-2, -1), (-0.1, 0.1), (2.5, 3), (4, 4.5)]

for (x_from, x_to) in chunks:
    x, y = fsolve(equations, (x_from, x_to))
    print('{:.6f}'.format(x), '{:.6f}'.format(y))
```
