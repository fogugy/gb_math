#### 1. Задание (в программе)
Нарисуйте график функции: y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
def f(x, k=1, a=0, b=0):
    return k * np.cos(x-a) + b
```

```python
x = np.linspace(0, 4*np.pi, 100)
plt.plot(x,f(x), color="red");
plt.plot(x,f(x, 2, a=0.3), color="green");
plt.plot(x,f(x, 2, b=1), color="blue");
plt.grid(True)
```

#### 4.1 Задание (в программе)
Решите систему уравнений:
y = x**2 – 1; 
exp(x) + x∙(1 – y) = 1

```python
point_count = 200
x = np.append(np.linspace(-2, 0.00001, point_count//2), np.linspace(0.00001, 3, point_count//2))
y = np.linspace(-2, 8, point_count)

x0 = np.array([0]*point_count)
y0 = np.array([0]*point_count)

y1 = x**2 - 1
y2 = np.exp(x)/x + 1 -1/x

plt.plot(x,y1, color="green")
plt.plot(x,y2, color="blue")

plt.plot(x,y0, color="red", linestyle='--', linewidth=1)
plt.plot(x0,y, color="red",linestyle='--', linewidth=1)

plt.grid(True)
```

```python
def equations(p):
    x, y = p
    return (x**2-1-y, exp(x)+x*(1-y)-1)

x1, y1 =  fsolve(equations, (-2, -1))
x2, y2 =  fsolve(equations, (2.5, 3))

print(f'P1: {x1}, {y1}')
print(f'P2: {x2}, {y2}')
```
